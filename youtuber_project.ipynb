```python
# we import the python libraries those we will need

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# now  we import the file which we'll dig into

data=pd.read_csv("C:/Users/LENOVO/Desktop/Kodlasam ders notları/Yeni klasör/ders notları 1/youtuber.csv")
data
```

```python
# setting Rank column as the new Index column
data.set_index("Rank",inplace=True)
# with 'inplace=True' command we make this setting as permanent
data
```

```python
# I realise miss spelling on a column name and i have to fix it for that ;
data.rename(columns={'Suscribers': 'Subscribers'}, inplace=True)
data
```

```python
# As we see from below , our data has no 'null' values so we don't need to check
data.info()
```

```python
# We need to change of the datatypes of numeric columns

# Remove commas
data["Subscribers"] = data["Subscribers"].str.replace(',', '') 

# Now, let's convert the values to integers, but we need to handle missing or invalid values
data["Subscribers"] = pd.to_numeric(data["Subscribers"], errors='coerce')

# "errors='coerce'" parameter will set invalid parsing to NaN (Not a Number)
# Now let's check what we did !     

data.info()
```

```python
# Here we set a list with the names of columns those we want to change the datatypes
columns_to_change = ["Visits", "Likes", "Comments"]
```

```python
data[columns_to_change] = data[columns_to_change].replace(',', '', regex=True)

data[columns_to_change] = data[columns_to_change].apply(pd.to_numeric, errors='coerce')
```

```python
data.info()
```

```python
# Now i want to eliminate all the youtubers from my list those don't have any visits or subscribers and rename my 
# dataframe as 'youtuber'
youtuber = data[(data['Visits'] != 0) & (data['Subscribers'] != 0)]
```

```python
youtuber_usa=youtuber[youtuber['Country']=='US']
youtuber_usa_subscribers=youtuber_usa.groupby('Categories')['Subscribers'].mean().sort_values(ascending=False).round(2)
youtuber_usa_subscribers
# Average Subscribers per Category of YouTubers whose country is USA 
```

```python
# Plotting mean subscribers
plt.figure(figsize=(10, 6))
plt.bar(youtuber_usa_subscribers.index, youtuber_usa_subscribers.values, color='blue', label='Mean Subscribers')

# Plotting mean visits
plt.bar(youtuber_usa_visits.index, youtuber_usa_visits.values, color='orange', label='Mean Visits')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Mean Values')
plt.title('Comparison of Mean Subscribers and Mean Visits by Category in USA')
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better readability
plt.legend()

# Displaying the plot
plt.tight_layout()
plt.show()
```
