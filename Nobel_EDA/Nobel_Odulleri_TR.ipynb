# 5.Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
# Define the categories for which you want to create charts
categories = ['Chemistry','Literature', 'Medicine', 'Peace', 'Physics', 'Economics']

# Create a grid of subplots: 3 rows, 2 columns
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

# Flatten the array of axes for easy iteration
axes = axes.flatten()

# Iterate through each category and plot line plots
for i, category in enumerate(categories):
    # Filter the DataFrame for the current category
    df_category = df[df['category'] == category]
    
    # Plot line plot for male data in the current category
    male_data = df_category[df_category['sex'] == 'Male']
    axes[i].plot(male_data['year'], male_data['age'], label='Male', color='blue', marker='o')
    
    # Plot line plot for female data in the current category
    female_data = df_category[df_category['sex'] == 'Female']
    axes[i].plot(female_data['year'], female_data['age'], label='Female', color='pink', marker='*')
    
    # Set plot labels and title
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Age')
    axes[i].set_title(f'Age Distribution in {category} Category')
    
    # Add a legend
    axes[i].legend()
    
    # Add a grid
    axes[i].grid(True, alpha=0.5)

# Remove any unused subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

**Bu soruda kurumların uç değerler olacağını düşündüğüm ve soruda da KİŞİLER dediği için sadece bireysel ödül kazananları aldım.**

```python
# Filter the DataFrame to include rows from the year 2000 onwards and 'Individual' laureate type
df_years3 = df[(df['year'] >= 2000) & (df['laureate_type'] == 'Individual')]

# Define the categories for which you want to create charts
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']

# Create a grid of subplots: 3 rows, 2 columns (but there are only 5 categories, so one subplot will remain empty)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

# Flatten the array of axes for easy iteration
axes = axes.flatten()

# Iterate through each category and plot boxplots
for i, category in enumerate(categories):
    # Filter the DataFrame for the current category
    df_category = df_years3[df_years3['category'] == category]
    
    # Plot the boxplot on the respective subplot
    ax = axes[i]
    df_category.boxplot(column='age', by='organization_country', ax=ax, patch_artist=True, showfliers=False, boxprops=dict(facecolor='lightblue'))
    
    # Set plot labels and title
    ax.set_xlabel('Country')
    ax.set_ylabel('Age')
    ax.set_title(f'Age Distribution in {category}')
    
    # Rotate x-axis labels by 90 degrees
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    
    # Remove the automatic title added by boxplot (set_title is specified above)
    ax.get_figure().suptitle('')
    
    # Add grid
    ax.grid(True, alpha=0.3)

# Remove any unused subplots
for j in range(len(categories), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
